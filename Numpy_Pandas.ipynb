```python
def nan_values(df):
    nan_val = df.isnull().sum()
    nan_val_per = (nan_val / len(df)) * 100
    nan_val_table = pd.concat([nan_val, nan_val_per], axis=1)
    nan_val_table = nan_val_table[nan_val_table.iloc[:,1] != 0]
    nan_val_table.rename(columns={0:"Fill rate",1:"Fill rate %"},inplace=True)
    return nan_val_table
```
